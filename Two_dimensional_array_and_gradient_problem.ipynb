Create a ndarray representing x at 0.1 intervals from -50 to 50, and a ndarray representing y for each element.

```python
import numpy as np
import matplotlib.pyplot as plt

x=np.arange(-50,50.1,0.1)
y=0.5*x+1
```

[Problem 2] Array combination
Combine the ndarrays created in Problem 1 into an array of the form (1001, 2). Make sure x is in the 0th column and y is in the 1st column.

```python
z=x.reshape(1001,1)
k=y.reshape(1001,1)
k.shape
a=np.concatenate([z,k], axis=1)
```

[Problem 3] Find the gradient
Find the gradient at each x-coordinate. Here, the gradient is calculated by dividing the change in y by the change in x.

```python
dx=a[1:,0]-a[:-1,0]
dy=a[1:,1]-a[:-1,1]
t=dy/dx
t
```

[Problem 5] Python functionalization
Please refer to the template to create a Python function that summarizes the problems 1 to 3.

Then, calculate the gradient of the following three equations at 0.1 intervals and output it with ndarray.

To calculate the sin of a trigonometric function for an ndarray, use np.sin ().

```python
def find_gradient(function, x_range=(-50, 50.1, 0.1)):
    """
    Calculate the gradient using the amount of change.

    Parameters
    ----------------
    function : function
      The function you want to find, the one that returns the ndarray of y when you put the ndarray of x.
    x_range : tuple
      Specify the range in the same way as the argument of np.arange ().
    dx : tuple 
      difference of x array
    dy : tuple
      difference of y array
    Returns
    ----------------
    xy : ndarray, shape(n, 2)
      A combination of x and y. n depends on x_range.
    gradient : ndarray, shape(n-1,)
      Function gradient. One factor is reduced to make a difference
    """
    x = np.arange(*x_range)
    y = function(x)
    xy=np.column_stack((x,y))
    dx=xy[1:,0]-xy[:-1,0]
    dy=xy[1:,1]-xy[:-1,1]
    gradient=dy/dx
    # Plot the function and the gradient
    fig, (ax1, ax2) = plt.subplots(nrows=2, sharex=True, figsize=(8, 8))
    ax1.plot(x, y)
    ax1.set_ylabel('y')
    ax1.set_title('Function and gradient')
    ax2.plot(x[:-1], gradient)
    ax2.set_xlabel('x')
    ax2.set_ylabel('Gradient')
    plt.show()
    return xy , gradient
def function1(x_range):
  y=x*x
  return y
def function2(x_range):
  y=2*x*x+2**x
  return y
def function3(x_range):
  y=np.sin(x**1/2)
  return y     
print("y=x*x 1 and gradient")
function1_xy, function1_gradient = find_gradient(function1)
print("y=2*x*x+2**x and gradient")
function2_xy, function2_gradient = find_gradient(function2)
print("y=np.sin(x**1/2) and gradient")
function3_xy, function3_gradient = find_gradient(function3)
```

Also, display the slope before and after taking the minimum y.

```python
import numpy as np

# Define the function
# Create an array for x values
x = np.arange(-50, 50.1, 0.1)

# Compute the y values for each x
def using_y_min(y):
  '''
  Function that calculates gradient using minimum of array
  
  Parameters:
  -----------
  min_y : np.array
  minimum number of array

  min_y_index : np.array
  index of minimum value

  slope_before : gradient before index of minimum value of array

  slope_after : gradient after index of minimum value of array
  
  Returns:
  --------
  Print minimum value and slope after and before
  '''
# Find the minimum value of y and its index
  min_y = np.min(y)
  min_y_index = np.argmin(y)

# Find the slope before and after taking the minimum y
  slope_before = (y[min_y_index] - y[min_y_index-1]) / (x[min_y_index] - x[min_y_index-1])
  slope_after = (y[min_y_index+1] - y[min_y_index]) / (x[min_y_index+1] - x[min_y_index])

# Print the results
  print("Minimum y value: {:.2f}".format(min_y))
  print("Index of minimum y value: {}".format(min_y_index))
  print("Slope before minimum y: {:.2f}".format(slope_before))
  print("Slope after minimum y: {:.2f}".format(slope_after))
using_y_min(function1(x))
using_y_min(function2(x))
using_y_min(function3(x))
```
